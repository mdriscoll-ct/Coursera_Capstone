```python
# Library - Numerical Library for Vector Data
import numpy as np
# Library - Numerical Library for Random # Generation
import random
# LIbrary - PANDA - Data Analytics 
import pandas as pd
# Library - Request Handling
import requests
```

```python
# Library - Ipython Import and Display Images
from IPython.display import Image 
from IPython.core.display import HTML 
from IPython.display import display_html

    
# Library to tranform json file into a pandas dataframe library
from pandas.io.json import json_normalize
```

```python
# Import Web Page for Canadian Postal Codes using Beautiful Soup and Display to import and display table

source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
    
tbl = str(soup.table)
display_html(tbl,raw=True)
```

```python
# Convert HTML table to Dataframe using Pandas
dfs = pd.read_html(tbl)
df=dfs[0]
df.head()
```

```python
# Drop Rows Where Borough is NOt ASSIGNED and Combine with Neighborhoods with the same Postal Code
df1 = df[df.Borough != 'Not assigned']
df1.head()
```

```python
# Combine Neighborhoods with the same Postal Code
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)
df2.head()
```

```python
# Replacing names neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

# Display Dataframe
df2
```

```python
# Display shape of dataframe
df2.shape
```

```python
# Import Lat and Long File
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
# Rename column to align with DF and Merge tables
lat_lon.rename(columns={'Postcode':'Postal Code'},inplace=True)
df3 = pd.merge(df2,lat_lon,on='Postal Code')
df3.head()
```

```python
# Import Folium Plotting Library and SKLEARN - Cluster and Plotting
!conda install -c conda-forge folium=0.5.0 --yes
import folium # plotting library

from sklearn.cluster import KMeans

import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# Extract Toronto Only
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
```

```python
df4
```

```python
# Plot DF Mapping Following NYC Example

map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
# Perform K-Means Clustering with 5 Segments
k=5
toronto_clustering = df4.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
```

```python
df4.insert(0, 'Cluster Labels', kmeans.labels_)
df4.head()
```

```python
# Plot Clusters
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

```

```python
# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
